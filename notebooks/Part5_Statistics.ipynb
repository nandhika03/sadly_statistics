# **Reliability and Validity**

## **Reliability**

- Reliabiltiy is the nothing but the value that measures the consistency of a test
- The test is said to be reliable only when it can be trusted and tested for consistency
- A test cannot be valid without being reliable
- Observed score = true score + error score
- We try to minimize the error value here

### **Test-retest reliability**

- Here, the test is checked if it gives same or similar results over a period of time with the same test and same test takers involved.

### **Parallel forms of reliability**

- We check the equivalnce or similarity of different forms of the same test
- A test can be taken over a period of 2 days in two different forms
- We should make sure that both these forms are similiar to one another

### **Internal consistency reliability**

- Here, we make sure that each item of the test measures the same dimension/construct or thing
- The individual test scores are correlated with the total test scores and then it is compared with the overall variability of the total test scores
- It is measured using different values like:
     - Cronbach's alpha
     - Spearman - Brown
     - KR 20 and KR 21 (Kuder Richardsons)
     - split - half reliability

### **Interrater reliability**

- Here, we check the consistency or similarity between the judges or people who rate a particular construct.

### **Coefficient of reliability**

- The co-efficient of reliability is always positive and the higher it is the more reliable the test is.

### **What happens when we struggle to establish reliability?**

- Standardize the testing methods
- Increase the sample size of observations
- Delete the unclear observations
- Moderate the level of easeness or difficulty of the test
- Decrease the effect of confounding variables

## **Validity**

- Here, we check if the test measures what it indents to measure

### **Content validity**

- Here, we check if the sample of the test items actually represent the universe of the testing construct or dimension.
- The questions in a history test should actually check the knowledge of the students in the domain of histoty
- Having a subject matter expert is crucial for content validity

### **Criterion validity**

- Here, we check if the criteria in the test actually goes well with what we are testing
- We may have both concurrent(present) and predictive(future) criterion validity
- We might want to check if the criteria is good by correlating it with judegs or expert's opinion

### **Construct validity**

- This is the most difficult validity to establish
- Here, we have to establish that our theory about some matter would actually (scientifically) be a good measurement for a particular behavior or attitude.
- This can take years to achieve

##### - The maximum level of validity = sqaure root (reliability co-efficient)
